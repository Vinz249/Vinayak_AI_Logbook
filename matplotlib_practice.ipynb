### Matplotlib

```python
x = [1, 2,2,1]
y = [3, 4,3,3]
plt.plot(x,y)
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Basic Plot')
plt.show()
```

Plot with good styling

```python
x = [1,2,3,4]
y = [50000,80000,60000,100000]
plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='s',label='2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Montly Sales Data Report')
plt.legend(loc = 'upper left',fontsize = 10)
plt.grid(color = 'grey', linestyle=':', linewidth = 1)
plt.xticks(x,['Jan','Feb','March','May'])
#plt.xlim(1,3)
#plt.ylim(0,100000)
plt.show()
```

Bar graph

```python
products= ['A','B','C','D','E']
sales = [50000,80000,60000,100000,20000]
plt.barh(products,sales,height =0.5, color='lightblue',label = 'Sales 2025')
plt.legend()
plt.xlabel('Products')
plt.ylabel('Annual Sales')
plt.title('Annual Sales Comparison Report')
plt.grid(color='grey',linestyle=':',linewidth=0.5)
plt.show()
```

Pie chart

```python
regions = ['North Delhi','East Delhi','West Delhi','South Delhi']
pollution = [500,600,300,1000]
plt.pie(pollution,labels=regions,autopct='%1.1f%%')
plt.title('Air Pollution Distribution in Delhi') 
plt.show()
```

Histogram

```python
marks = [50,75,80,92,5,25,85,54,65,76,45,67,72,71,77,99,64,61,76,6]
plt.hist(marks,bins=6,color='purple',edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Score distribution of Students')
plt.show()
```

Scatter plot

Subplot: quick and easy and not recommended way

```python
x=[1,3,5,7,9]
y=[100,200,300,400,500]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar graph')

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line Graph')
plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar Graph')
plt.show()
```

Recommended way to do subplot()

```python
fig, ax = plt.subplots(1,2,figsize=(8,4))
ax[0].plot(x,y,label='Graph Data')
ax[0].set_title('Line Graph')
ax[0].set_xlabel('X input')
ax[0].set_ylabel('Y output')
ax[0].legend()
ax[1].bar(x,y,label='Graph Data')
ax[1].set_title('Bar Graph')
ax[1].set_xlabel('X input')
ax[1].set_ylabel('Y output')
ax[1].legend()
fig.suptitle('Comparison of Line Graph and Bar Graph')

plt.tight_layout()
plt.show()
```
